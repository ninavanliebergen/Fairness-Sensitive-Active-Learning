# Synthetical dataset

4 groups:
1. South Netherlands - high income (#2000)
2. South Netherlands - low income (#5000)
3. North Netherlands - high income (#2000)
4. North Netherlands - Low income (#1000)

Group 3 has a historical bias towards woman: They are classified as more frauduleus than man. This is because of the proxy value 'Occupation'. People with the occupation Teacher are classifier more as frauduleus in the north of netherlands when having a high income. 

```python
import pandas as pd
```

```python
from numpy.random import \
    rand as np_rand, \
    choice as np_choice, \
    normal as np_normal, \
    seed as np_seed

from numpy import \
    vstack as np_vstack, \
    hstack as np_hstack, \
    transpose as np_transpose, \
    mean as np_mean
    

from sklearn.preprocessing import MinMaxScaler

def get_binary(lst, threshold):
    tmp = []
    for value in lst:
        if value >= 0:
            tmp.append(1)
        else:
            tmp.append(0)    
    return tmp
    

def get_samples_NL_HOOG(n_samples, seed=1102, include_hidden=False):

    np_seed(seed)
    
    # NL HOOG inkomen
    x0 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[1, 0])
    x1 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[1, 0])
    
    # beroep en aantal jaar in dienst
    x2 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    x3 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    
    # sensitieve attribuut
    s = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    
    # tranfsorm x2 tot proxy van sensitieve attribuut
    s1 = s*2 + x2
    tmp = s1
    s1 = []
    
    # for the proxy: if s and x are both 1, than very high chance of 0
    for element in tmp:
        if element == 4:
            ls_value = np_choice(a=[1, 0], size=1, replace=True, p=[0, 1])
            s1.append(ls_value[0])
        if element == 3:
            ls_value = np_choice(a=[1, 0], size=1, replace=True, p=[0.1, 0.9])
            s1.append(ls_value[0])
        if element == 2:
            ls_value = np_choice(a=[1, 0], size=1, replace=True, p=[0.5, 0.5])
            s1.append(ls_value[0])
        if element == 1:
            ls_value = np_choice(a=[1, 0], size=1, replace=True, p=[0.9, 0.1])
            s1.append(ls_value[0])  
        if element == 0:
            ls_value = np_choice(a=[1, 0], size=1, replace=True, p=[1, 0])
            s1.append(ls_value[0])
        
    # target value = y
    y = x1 + s1 + np_normal(0, 0.5, n_samples)
    y = (y > np_mean(y)) + 0

    # hidden is the proxy
    if include_hidden:
        X1 = np_transpose(np_vstack((x0, x1, s1, x3, y, s)))
    else:
        X1 = np_transpose(np_vstack((x0, x1, s1, x3, y, s)))
     
#     scaler = MinMaxScaler()
#     X = scaler.fit_transform(X1)
    X = X1
    
    X = pd.DataFrame(X, columns=['State', 'Income', 'Occupation', 'Time employed', 'Fraud', 'Gender'])  # 1st row as the column names
    return X

def get_samples_NL_LAAG(n_samples, seed=1103, include_hidden=False):

    np_seed(seed)
    
    # NL LAAG inkomen
    x0 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[1, 0])
    x1 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[0, 1])
    
    # beroep en aantal jaar in dienst
    x2 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    x3 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    
    # sensitieve attribuut
    s = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
        
    # target value = y
    y = x1 + x2 + np_normal(0, 0.5, n_samples)
    y = (y > np_mean(y)) + 0

    # hidden is the proxy
    if include_hidden:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
    else:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
        
#     scaler = MinMaxScaler()
#     X = scaler.fit_transform(X1)
    X = X1
    X = pd.DataFrame(X, columns=['State', 'Income', 'Occupation', 'Time employed', 'Fraud', 'Gender'])  # 1st row as the column names
    return X

def get_samples_ZN_HOOG(n_samples, seed=1104, include_hidden=False):

    np_seed(seed)
    
    # NL LAAG inkomen
    x0 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[0, 1])
    x1 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[1, 0])
    
    # beroep en aantal jaar in dienst
    x2 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    x3 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    
    # sensitieve attribuut
    s = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
        
    # target value = y
    y = x1 + x2 + np_normal(0, 0.5, n_samples)
    y = (y > np_mean(y)) + 0

    # hidden is the proxy
    if include_hidden:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
    else:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
     
      
#     scaler = MinMaxScaler()
#     X = scaler.fit_transform(X1)
    X = X1
    
    X = pd.DataFrame(X, columns=['State', 'Income', 'Occupation', 'Time employed', 'Fraud', 'Gender'])  # 1st row as the column names
    return X

def get_samples_ZN_LAAG(n_samples, seed=1105, include_hidden=False):

    np_seed(seed)
    
    # NL LAAG inkomen
    x0 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[0, 1])
    x1 = np_choice(a=[0, 1], size=n_samples, replace=True, p=[0, 1])
    
    # beroep en aantal jaar in dienst
    x2 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    x3 = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
    
    # sensitieve attribuut
    s = np_choice(a=[0,1], size=n_samples, replace=True, p=[0.5, 0.5])
        
    # target value = y
    y = x1 + x2 + np_normal(0, 0.5, n_samples)
    y = (y > np_mean(y)) + 0

    # hidden is the proxy
    if include_hidden:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
    else:
        X1 = np_transpose(np_vstack((x0, x1, x2, x3, y, s)))
     
#     scaler = MinMaxScaler()
#     X = scaler.fit_transform(X1)
    X = X1
    
    X = pd.DataFrame(X, columns=['State', 'Income', 'Occupation', 'Time employed', 'Fraud', 'Gender'])  # 1st row as the column names
    return X
```

# Calculate the fairness and unfairness on all the datapoints of the total dataset

```python
S = merged_df['Gender']
merged_df = merged_df.drop("Gender", axis=1)
merged_df = merged_df.drop("Group", axis=1)
y = merged_df['Fraud']
merged_df = merged_df.drop("Fraud", axis=1)
```

```python
merged_df
```

```python
import fairlearn
from fairlearn.metrics import demographic_parity_difference, demographic_parity_ratio, equalized_odds_difference, equalized_odds_ratio

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def evaluate_fairness(X, y, S, num_splits=30):
    accuracy_scores = []
    dpd_scores = []
    dpr_scores = []
    eod_scores = []
    eor_scores = []

    for _ in range(num_splits):
        # Split the dataset into training and testing sets
        X_train, X_test, y_train, y_test, S_train, S_test = train_test_split(
            X, y, S, test_size=0.3, random_state=None)  # Remove random_state
        
        # Initialize the Logistic Regression model
        model = LogisticRegression()

        # Train the model on the training data
        model.fit(X_train, y_train)

        # Make predictions on the test data
        y_pred = model.predict(X_test)

        # Calculate the accuracy of the model
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_scores.append(accuracy)

        # Calculate the fairness metrics for the current split
        dpd = demographic_parity_difference(y_test, y_pred, sensitive_features=S_test)
        dpr = demographic_parity_ratio(y_test, y_pred, sensitive_features=S_test)
        eod = equalized_odds_difference(y_test, y_pred, sensitive_features=S_test)
        eor = equalized_odds_ratio(y_test, y_pred, sensitive_features=S_test)

        dpd_scores.append(dpd)
        dpr_scores.append(dpr)
        eod_scores.append(eod)
        eor_scores.append(eor)

    # Calculate mean and standard deviation of the metrics
    mean_accuracy = np.mean(accuracy_scores)
    std_accuracy = np.std(accuracy_scores)

    mean_dpd = np.mean(dpd_scores)
    std_dpd = np.std(dpd_scores)

    mean_dpr = np.mean(dpr_scores)
    std_dpr = np.std(dpr_scores)

    mean_eod = np.mean(eod_scores)
    std_eod = np.std(eod_scores)

    mean_eor = np.mean(eor_scores)
    std_eor = np.std(eor_scores)

    results = {
        "Mean Accuracy": mean_accuracy,
        "Accuracy Standard Deviation": std_accuracy,
        "Mean Demographic Parity Difference": mean_dpd,
        "Demographic Parity Difference Standard Deviation": std_dpd,
        "Mean Demographic Parity Ratio": mean_dpr,
        "Demographic Parity Ratio Standard Deviation": std_dpr,
        "Mean Equalized Odds Difference": mean_eod,
        "Equalized Odds Difference Standard Deviation": std_eod,
        "Mean Equalized Odds Ratio": mean_eor,
        "Equalized Odds Ratio Standard Deviation": std_eor
    }

    return results

# Example usage
# X = ...  # Features
# y = ...  # Target labels
# S = ...  # Sensitive features
results = evaluate_fairness(merged_df.values, y.values, S.values)
for metric, value in results.items():
    print(metric + ":", value)
    print("\n")

```
